## RandomUser API Example:

```python
import pandas as pd
from randomuser import RandomUser #generates random user data using the randomuser.me API.
```

```python
user=RandomUser

#Make a list of 15 random users
user_list= RandomUser.generate_users(15)
user_list
```

```python
#Make a list of 15 random users with their first name and date of birth
for user in user_list:
    print(user.get_first_name(), ' ', user.get_dob())
```

```python
#Write a function that will make a user information table and return a dataframe.
def get_users():
    users=[] #empty list

    for user in user_list:
        #Add element to end of list
        users.append({'Name':user.get_full_name(), 'Email':user.get_email(), 'Phone Number':user.get_phone(), 'State':user.get_state()}) #Append a dictionary to the empty list(users)

    return pd.DataFrame(users) #output is a pandas dataframe

get_users()
```

## Fruitvice API:

```python
import requests #Send HTTP requests using Python
import json #Convert datastructures to JSON strings
```

```python
url= ("https://fruityvice.com/api/fruit/all")
response=requests.get(url)

results = json.loads(response.text) #Parse a valid JSON string and convert it into a Python Dictionary
#deserialize the text attribute of the request object

df1=pd.DataFrame(results) #convert json data into a pandas dataframe
df1
```

The nutritions column has several subcolumns so the data will need to be normalized (flattened)

```python
df2=pd.json_normalize(results) #Normalize semi-structured JSON data into a flat table
df2
```

What is the genus and order of blackberry?

```python
blackberry=df2.loc[df2['name']=='Blackberry'] #.loc to select Blackberry from name column
(blackberry.iloc[0]['genus']), (blackberry.iloc[0]['order']) #.iloc to select index 0 of genus and order column
```

How many calories are in dragonfruit?

```python
dragonfruit_calories=df2.loc[df2['name']=='Dragonfruit']
dragonfruit_calories.iloc[0]['nutritions.calories']
```
