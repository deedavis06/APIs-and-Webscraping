## Introduction

We are going to make a candlestick chart providing information about stock prices using the Matplotlib Finance API.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd 
import mplfinance as mpl #matplotlib utilities for financial data
import yfinance as yf #data from yahoo finance
```

We will be looking at the last 5 years of Apple Inc. stocks.

```python
#.Ticker refers to the ticker/stock symbol used to identify publicly traded stocks (e.g. APPL is the ticker for Apple Inc.)
#.history returns all information about Apple Inc. from the last 5 years
df=yf.Ticker('AAPL').history(period='5y')
print(df)  
```

```python
#Get the column names
df.columns
```

```python
mpl.plot(df, type='candle')
```

Now thats a lot of data! Lets just look at the data from January 1,2023 to today (February 16, 2023)

```python
df=df.loc['2023-01-01':]
#Plot a candle chart, a volume profile and a title
mpl.plot(df, type='candle', volume=True,title='Apple (USD) from January 1, 2023')
```

```python
#Examine the time frame from 1/1/23 to today (2/16/23)
df=df.loc['2023-01-01':]
#Add a simple moving average (mav) at 5 and 10 days
mpl.plot(df, type='candle', mav=(5,10), volume=True,title='Apple (USD) from January 1, 2023')
```
